# Explore the Crimes conducted for Delhi

```python
delhi_total_rape = df.loc[df['Area_Name']=='Delhi']
delhi_victim_2010_total = delhi_total_rape [(delhi_total_rape['Year']==2010) & (delhi_total_rape['Subgroup']=='Total Rape Victims')]
delhi_victim_2010_total_incest_rape = delhi_total_rape [(delhi_total_rape['Year']==2010) & (delhi_total_rape['Subgroup']=='Victims of Incest Rape')]
```

```python
#Plotting age breakup of victims
ax = delhi_victim_2010_total[['Victims_Upto_10_Yrs','Victims_Between_10-14_Yrs','Victims_Between_14-18_Yrs','Victims_Between_18-30_Yrs','Victims_Between_30-50_Yrs']].plot(kind='bar',legend=True, title = 'Age Breakup of rape victims (Delhi.!!)')
ax.set_ylabel("No of Victims", fontsize=12)
ax.set_xticklabels([]);
ax = delhi_victim_2010_total_incest_rape[['Victims_Upto_10_Yrs','Victims_Between_10-14_Yrs','Victims_Between_14-18_Yrs','Victims_Between_18-30_Yrs','Victims_Between_30-50_Yrs']].plot(kind='bar',legend=True, title = 'Age Breakup of Incest rape victims (Delhi.!!)')
ax.set_ylabel("No of Victims", fontsize=12)
ax.set_xticklabels([]);
```

# Total No. of Rapes in the year 2010

```python
rape_2010_total = df[(df['Year']==2010) & (df['Subgroup']== 'Total Rape Victims')]
ax1 = rape_2010_total['Victims_of_Rape_Total'].plot(kind='barh',figsize=(20, 15))
ax1.set_xlabel("Number of rape victims (2010)", fontsize=25)
ax1.set_yticklabels(rape_2010_total['Area_Name']);
```

Here we can see that Madhya Pradesh had the most no. of rape incidents for the year 2010!

We need to find out the reason behind this crime.

# Correlation between all kinds of Age Groups with Year

```python
# Find out the correlation between variables
corr=df.corr()
corr
```

```python
# visualize correlation between the variables using heatmap
plt.figure(figsize=(8,8))
sns.heatmap(corr, xticklabels = corr.columns.values, yticklabels = corr.columns.values,annot=True);
```

Here we can see that 97% of the Rape cases reported are for victims whose age between 18 to 30.

Year is less correlated 

```python
# Mean of the Total rape cases
df_mean = df.Victims_of_Rape_Total.mean()
print(df_mean)
```

```python
# Check out the states where the Total Rapes are greater than MEAN
df_total = df[df.Victims_of_Rape_Total>362]
df_total.head()
```

```python
# Visualize those states where the Rapes are more than mean
plt.figure(figsize=(24,5))
plt.title('Count of States, where the Rapes are more than Mean Value of Rape!');
sns.countplot(df_total.Area_Name);
plt.xticks(rotation = 60);
```

```python
# Let's explore the recieved Data Frame and perform drilling analysis on that!
data_frame_to_drill = df_total.copy()
```

```python
data_frame_to_drill.describe()
```

```python
data_frame_to_drill.shape
```

```python
# Check out the sum of Victims which are upto 10 Years in age only
data_frame_to_drill.Victims_Upto_10_Yrs.sum()
```

```python
# Let's visualize the Victims between 14 to 18 years out of Total Rape cases
plt.figure(figsize=(24,8))
sns.barplot(x = data_frame_to_drill.Area_Name, y = data_frame_to_drill['Rape_Cases_Reported'], color = "red")
upto_18 = sns.barplot(x = data_frame_to_drill.Area_Name, y = data_frame_to_drill['Victims_Between_14-18_Yrs'], color = "#a8ddb5")
topbar = plt.Rectangle((0,0),1,1,fc="red", edgecolor = 'none')
upto_18 = plt.Rectangle((0,0),1,1,fc='#a8ddb5',  edgecolor = 'none')
l = plt.legend([upto_18, topbar], ['Victims between 14 to 18 years of Age', 'Total Rape Cases'], loc=1, ncol = 2, prop={'size':16})
l.draw_frame(False)
sns.despine(left=True)
```

```python
# Let's visualize the Victims between 18 to 30 years out of Total Rape case
plt.figure(figsize=(24,8))
sns.barplot(x = data_frame_to_drill.Area_Name, y = data_frame_to_drill['Rape_Cases_Reported'], color = "red")
upto_30 = sns.barplot(x = data_frame_to_drill.Area_Name, y = data_frame_to_drill['Victims_Between_18-30_Yrs'], color = "#fe9929")
topbar = plt.Rectangle((0,0),1,1,fc="red", edgecolor = 'none')
upto_30 = plt.Rectangle((0,0),1,1,fc='#fe9929',  edgecolor = 'none')
l = plt.legend([upto_30, topbar], ['Victims between 18 to 30', 'Total Rape Cases'], loc=1, ncol = 2, prop={'size':16})
l.draw_frame(False)
sns.despine(left=True)
```

```python
plt.figure(figsize=(24,8))
sns.boxplot(data_frame_to_drill.Victims_Upto_10_Yrs,data_frame_to_drill.Area_Name,data = data_frame_to_drill);
plt.title('Spread of Rapes performed on Children upto 10 Years!');
```

We can see that the States Madhya Pradesh and Maharashtra both are closed to each other in the no. of rape crimes
